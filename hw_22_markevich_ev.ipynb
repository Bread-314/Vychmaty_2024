```python
from sympy import fibonacci

def di(a, b, f):
    eps = 1e-5
    while (b - a) > eps:
        c = (a + b)/2
        if f(c - eps) < f(c + eps):
            a, b = a, c
        else:
            a, b = c, b
    return (a+b)/2

def gold(a, b, f):
    N = 20
    x1 = a + ((b - a) * int(fibonacci(N - 2)) / int(fibonacci(N)))
    x2 = a + ((b - a) * int(fibonacci(N - 1)) / int(fibonacci(N)))
    f1 = f(x1)
    f2 = f(x2)
    for i in range(N, 1, -1):
        if f1 > f2:
            a, x1 = x1, x2
            x2 = b - (x1 - a)
            f1, f2 = f2, f(x2)
        else:
            b, x2 = x2, x1
            x1 = a + (b - x2) 
            f2, f1 = f1, f(x1)

    return (x1 + x2)/2

```

```python
def f1(t):
    return (t-5)*np.exp(t)

def f2(t):
    return t**4 / 4 - 2 * t**2
```

```python
print('min of f1 by dih- ', di(0, 5, f1))
print('min of f2 by dih - ', di(0, 5, f2))
```

```python
print('min of f1 by fib - ', gold(0, 5, f1))
print('min of f2 by fib - ', gold(0, 5, f2))
```

# Задание 3

```python
from scipy.optimize import minimize

def g(x):
    return 20 + x[0]**2 + x[1]**2 - 10 * np.cos(2 * np.pi * x[0]) - 10 * np.cos(2 * np.pi* x[1])

def anti_g(x):
    return -1 * g(x)


x_mesh = np.linspace(-5, 5, 100, endpoint=True)
y_mesh = np.linspace(-5, 5, 100, endpoint=True)

X, Y = np.meshgrid(x_mesh, y_mesh)

minis = []
maxis = []

eps = 1e-1

for x0, y0 in zip(X.ravel(), Y.ravel()):
    start = [x0, y0]
    goal_min = minimize(g, start, method='COBYLA')
    goal_max = minimize(anti_g, start, method='COBYLA')
    if not minis:
        minis.append(goal_min.x)
    else:
        if goal_min.success:
            if np.max(np.linalg.norm((np.array(minis) - np.array(goal_min.x)), axis = 1)) > eps:
                minis.append(goal_min.x)

    if not maxis and np.linalg.norm(np.array(goal_max.x), 2) <= 10:
        maxis.append(goal_max.x)
    else:
        if goal_max.success and np.linalg.norm(np.array(goal_max.x), 2) <= 10:
            if np.max(np.linalg.norm((np.array(maxis) - np.array(goal_max.x)), axis = 1)) > eps:
                maxis.append(goal_max.x)
    

minis = np.array(minis)
maxis = np.array(maxis)
plt.figure(figsize=(20, 10))
plt.scatter(minis[:, 0], minis[:, 1], label = 'min', color = 'blue')
plt.scatter(maxis[:, 0], maxis[:, 1], label = 'max', color = 'red')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.optimize import minimize_scalar

def St_T(x):
    return (x[0]**4 + x[1]**4 - 16*(x[0]**2 + x[1]**2) + 5*(x[0] + x[1]))/2

def grad_St_T(x):
    return np.array([4 * x[0]**3 - 32 * x[0] + 5, 4 * x[1]**3 - 32 * x[1] + 5])/2


def draw_function_and_grad(points):
    # Создаем сетку для тепловой карты
    x = np.linspace(-4, 1, 500)
    y = np.linspace(-4.5, -0.5, 500)
    X, Y = np.meshgrid(x, y)
    Z = St_T([X, Y])

    # Строим тепловую карту
    plt.figure(figsize=(15, 12))
    plt.contourf(X, Y, Z, levels=50, cmap='viridis')
    plt.colorbar(label='St_T(x, y)')

    # Наносим стрелки между точками
    for i in range(len(points) - 1):
        start = points[i]
        end = points[i + 1]
        dx, dy = end[0] - start[0], end[1] - start[1]
        plt.quiver(start[0], start[1], dx, dy, angles='xy', scale_units='xy', scale=1, color='red', width=0.003)

    # Отображаем точки
    points_x = [p[0] for p in points]
    points_y = [p[1] for p in points]
    plt.scatter(points_x, points_y, color='white', edgecolor='black', label='Точки', s = 20)

    plt.title('Тепловая карта функции St_T с последовательными точками')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

def nonlinear_conjugate_gradient_method(x0, method):
    x = x0
    its = 0
    r_old = -1 * grad_St_T(x)
    eps = 1e-6
    s = r_old
    steps = [np.array(x)]
    while its < 1000 and np.linalg.norm(r_old) > eps:
        x = x + minimize_scalar(lambda a: St_T(x + a * s)).x * s
        steps.append(x)
        r_new = -1 * grad_St_T(x)
        if method == 1:
            betta = (r_new @ r_new)/(r_old @ r_old)
        elif method == 2:
            betta = (r_new @ (r_new - r_old))/(r_old @ r_old)
        elif method == 3:
            betta = (r_new @ (r_new - r_old))/(s @ (r_new - r_old))
        elif method == 4:
            betta = (r_new @ r_new)/((-1 * s) @ r_old)
        elif method == 5:
            betta = (r_new @ (r_new - r_old))/ ((-1 * s) @ r_old)
        elif method == 6:
            betta = (r_new @ r_new)/(s @ (r_new - r_old))
        else:
            betta = (r_new - r_old - 2 * s * ((r_new - r_old) @ (r_new - r_old)) / (s @ r_new - r_old)) @ (r_new / (s @ (r_new - r_old)))

        s = r_new + betta * s
        its += 1
        r_old = r_new
    draw_function_and_grad(steps)
    return f"начальная точка - {str(x0):<13}, минимум - {str(x):<25}, количество иттераций - {its:<5}"

x1 = [0.1, -1]
x2 = [0.1, -1.1]
x3 = [0, -1]
```

```python
print("Метод Флетчер, Ривз")
print(nonlinear_conjugate_gradient_method(x1, 1))
print(nonlinear_conjugate_gradient_method(x2, 1))
print(nonlinear_conjugate_gradient_method(x3, 1))
```
