```python
import numpy as np
from scipy.linalg import block_diag

import matplotlib.pyplot as plt


n = 5

# Диагональный блок
a = np.zeros((n-1, n-1))

idx = np.arange(n-1)

a[idx, idx] = -4
a[idx[:-1], idx[:-1]+1] = 1
a[idx[1:], idx[1:]-1] = 1

# собираем блочно-диагональную `m`
m = block_diag(a, a, a, a, a)

# заполняем "крылья"
idx = np.arange(m.shape[0])

m[idx[:-n+1], idx[:-n+1] + n-1] = 1
m[idx[n-1:], idx[n-1:] - n+1] = 1

with np.printoptions(linewidth=99):
    print(m)
```

```python
#Третий метод
# !pip install numba
# !pip install "numpy<2.0"
from numba import njit, objmode, prange

@njit()
def system_numba(n: int, alpha: float):

    A = np.zeros((n, n))
    f = np.zeros((n, 1))


    for i in range(n):
        A[i,i] = 2.
        if i > 0:
            A[i,i-1] = -1 + alpha
        if i < n - 1:
            A[i,i+1] = -1 - alpha

    f = np.zeros(n)
    f[0] = 1 - alpha
    f[n-1] = 1 + alpha

    return (A, f)

@njit()
def gauss_solve_numba(A,f):

    x = np.copy(f)

    n = len(f)
    # Elimination phase
    for k in range(0,n-1):
        for i in range(k+1,n):
            if A[i,k] != 0.0:
                #if not null define λ
                lam = A[i,k]/A[k,k]
                #we calculate the new row of the matrix
                A[i,k+1:n] = A[i,k+1:n] - lam*A[k,k+1:n]
                #we update vector b
                x[i] = x[i] - lam*x[k]
                # backward substitution
    for k in range(n-1,-1,-1):
        x[k] = (x[k] - np.dot(A[k,k+1:n],x[k+1:n]))/A[k,k]
    return x

@njit()
def jacobi_solve_numba(A, f, eps = 1e-6, MAX_ITER = 100000, x_prev = None):

    n = f.shape[0]
    if x_prev == None:
        x_prev = np.ones(f.shape)*2 # нулевое начальное приближение - плохо!
    x_curr = x_prev*1
    k = 0
    for k in range(1, MAX_ITER + 1):
        x_prev[:] = x_curr[:] # Это важно, да

        for i in range(n):
            x_curr[i] = 1./A[i, i]*(f[i] - A[i, 0:i] @ x_prev[0:i] - A[i, i+1:] @ x_prev[i+1:])
        if np.sum((x_curr - x_prev)**2) < eps**2:
            break
    if k == MAX_ITER:
        print("Didn't converge!Q!11!!1")
    return x_curr

@njit()
def zeidel_solve_numba(A, f, eps = 1e-6, MAX_ITER = 100000, x_prev = None):

    n = f.shape[0]
    if x_prev == None:
        x_prev = np.ones(f.shape)*2 # нулевое начальное приближение - плохо!
    x_curr = x_prev*1
    k = 0
    for k in range(1, MAX_ITER + 1):
        x_prev[:] = x_curr[:] # Это важно, да

        for i in range(n):
            x_curr[i] = 1./A[i, i]*(f[i] - A[i, 0:i] @ x_curr[0:i] - A[i, i+1:] @ x_prev[i+1:])
        if np.sum((x_curr - x_prev)**2) < eps**2:
            break
    if k == MAX_ITER:
        print("Didn't converge!Q!11!!1")
    return x_curr

@njit(parallel=True)
def series_timed_numba(alpha, ns, func):

    Ts = np.zeros(ns.shape)

    for i in prange(len(ns)): # только числовые циклы, prange для распараллеливания НЕЗАВИСИМЫХ циклов
                              # (т.е. шаг цикла не должен опираться на вычисления с предыдущих шагов)

        A, f = system_numba(ns[i], alpha)


        with objmode(start='f8'):
            start = time.perf_counter()

        _ = func(A, f)

        with objmode(seconds='f8'):
            seconds = time.perf_counter() - start

        Ts[i] = seconds # Убрал append

    return Ts
```

## Задание 2. (Метод Шульца (минимальных невязок) для вычисления обратной матрицы)

```python
import numpy as np

def matrix_inv(A, X_pred, m=10, eps=1e-6, max_iter=10000):
    E = np.eye(A.shape[0])
    Psi = E - A @ X_pred
    X_prev = X_pred
    X_curr = X_prev.copy()
    iter_num = 0

    while np.linalg.norm(Psi) > eps:
        iter_num += 1
        if iter_num > max_iter:
            print('bad, not good')
            return X_curr

        Psi_sum = E + Psi
        tmp = np.copy(Psi)
        for i in range(1, m - 1):
            
            tmp = tmp @ Psi
            Psi_sum += tmp
            
        X_curr = X_prev @ Psi_sum
        X_prev = X_curr
        Psi = E - A @ X_curr

    return X_curr
```

2. Возьмем $ k = j, l = i$, $a'_{ij} = a_{ji}\sin{\theta} + a_{ii}\cos{\theta}=0\Rightarrow \sin{\theta} = \sqrt{\frac{a_{ii}^2}{a_{ii}^2+a_{ji}^2}},\ \cos{\theta} = \sqrt{\frac{a_{ji}^2}{a_{ii}^2+a_{ji}^2}}$
3. Будем последовательно действовать матрицами поворота Гивенса на матрицу A и обнулять элементы под диагональю, пока не получим верхнетреугольную матрицу. Поскольку матрица Гивенса ортогональна, то их произведение также ортогональная матрица, поэтому получаем QR-разложение.

## Задача 5. SVD для разреженных матриц и МНК.

```python
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np
import time


# svd разложение 

def pseudoinverse_svd(A, eps=1e-2):
    U, S, Vh = np.linalg.svd(A, full_matrices=False)
    pseudo_S = np.zeros((A.shape[1], A.shape[0]))
    for i in range(len(S)):
        if S[i] < eps:
            continue
        pseudo_S[i, i] = 1 / S[i]
    return Vh.T @ pseudo_S @ U.T

def MSE_solve_SVD(A,b):
    return pseudoinverse_svd(A)@b

#qr

def gauss_solve(A,f):

    x = np.copy(f)

    n = len(f)
    # Elimination phase
    for k in range(0,n-1):
        for i in range(k+1,n):
            if A[i,k] != 0.0:
                #if not null define λ
                lam = A[i,k]/A[k,k]
                #we calculate the new row of the matrix
                A[i,k+1:n] = A[i,k+1:n] - lam*A[k,k+1:n]
                #we update vector b
                x[i] = x[i] - lam*x[k]
                # backward substitution
    for k in range(n-1,-1,-1):
        x[k] = (x[k] - np.dot(A[k,k+1:n],x[k+1:n]))/A[k,k]

    return x

def MSE_solve_QR(A,b):
    Q,R = np.linalg.qr(A,'complete')
    f = Q.T@b
    return gauss_solve(R,f)

def singular_values_attenuation_in_sparse(shape:tuple,density, func):
    '''
    Исследует убывание сингулярных чисел в разреженных матрицах
    размера shape и с плотностью density (от 0 до 1)
    '''
    sing_values_sum = np.zeros(min(shape)- 1 )

    sample_size = 100
    Ts = np.zeros(len(density))
    
    for j, dens in enumerate(density):
        T_sum = 0
        for i in range(sample_size):
            sparse = sp.sparse.rand(shape[0], shape[1],
                    density=dens,
                    random_state=42).toarray()

            b = np.random.rand(shape[0])
            start = time.time()
            _ = func(sparse,b)
            seconds = time.time() - start
            T_sum += seconds/sample_size
        
        Ts[j] = T_sum

        
    return Ts
```

```python
size = 10

sparse = sp.sparse.rand(size, size,
                    density=0.5,
                    random_state=42).toarray()

b = np.random.rand(size)

print(MSE_solve_SVD(sparse,b))
print(MSE_solve_QR(sparse,b))
print(np.linalg.pinv(sparse)@b)

density = np.linspace(0,1,100)
Ts_SVD = 10**6*singular_values_attenuation_in_sparse((size,size),density, MSE_solve_SVD)
Ts_QR = 10**6*singular_values_attenuation_in_sparse((size,size),density, MSE_solve_QR)
plt.plot(density,Ts_SVD,color = 'blue',label = 'svd')
plt.plot(density,Ts_QR,color = 'orange', label = 'qr')


plt.legend()
```
