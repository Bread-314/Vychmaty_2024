```python
#Exersice_6
#1
N = int(1e3)
M = 50
eigs = []
for i in range(M):
    norm_mat = np.random.normal(loc=0, scale = 0.7, size=(N,N))
    norm_mat += norm_mat.T
    eigenvalues, _ = np.linalg.eig(norm_mat)
    eigs.append(eigenvalues)
eigs = np.array(eigs)
```

```python
max_eig = np.max(eigs)
min_eig = np.min(eigs)

bins = np.linspace(min_eig, max_eig, 41)
counts = []
for i in range(M):
    cnt, bin_gaps = np.histogram(eigs[i], bins = bins)
    counts.append(cnt)
counts = np.array(counts)
final_data = np.mean(counts, axis = 0)
min_data = np.min(counts, axis=0)
max_data = np.max(counts, axis=0)

bin_centers = 0.5 * (bin_gaps[1:] + bin_gaps[:-1])

plt.figure(figsize=(9,7))
plt.bar(bin_centers, final_data, width=bin_gaps[1] - bin_gaps[0])
plt.errorbar(bin_centers, final_data, yerr=(final_data-min_data, max_data-final_data), fmt='.r', capsize=5)
plt.show()
```

```python
#2
N = int(5e2)
m = 100
rho = np.array([])
for i in range(m):
    norm_mat = np.random.normal(loc=0, scale = 0.7, size=(N,N))
    norm_mat += norm_mat.T
    eigenvalues, _ = np.linalg.eig(norm_mat)
    eigenvalues = np.sort(eigenvalues)
    rho = np.concatenate((rho, np.diff(eigenvalues)), axis = 0)
```

```python
#4
N = int(1e3)
M = 100
eigs = []
for i in range(M):
    x = np.random.randint(-1000, 1000, size=N*N)
    y = np.random.randint(-1000, 1000, size=N*N)
    mat = (x+1j*y).reshape((N, N))
    mat += mat.conjugate().T
    eigenvalues, _ = np.linalg.eig(mat)
    eigs.append(eigenvalues)
eigs = np.array(eigs)
```

```python
eigs = np.real(eigs)

max_eig = np.max(eigs)
min_eig = np.min(eigs)

bins = np.linspace(min_eig, max_eig, 41)
counts = []
for i in range(M):
    cnt, bin_gaps = np.histogram(eigs[i], bins = bins)
    counts.append(cnt)
counts = np.array(counts)
final_data = np.mean(counts, axis = 0)
min_data = np.min(counts, axis=0)
max_data = np.max(counts, axis=0)

bin_centers = 0.5 * (bin_gaps[1:] + bin_gaps[:-1])

plt.figure(figsize=(9,7))
plt.bar(bin_centers, final_data, width=bin_gaps[1] - bin_gaps[0])
plt.errorbar(bin_centers, final_data, yerr=(final_data-min_data, max_data-final_data), fmt='.r', capsize=5)
plt.show()
```
